# Analyzing Data from DataFrame

#### What is total cases & deaths ?

```python
total_cases = covid_df.new_cases.sum()
total_deaths = covid_df.new_deaths.sum()
type(total_deaths)
```

```python
print('Total cases: {}, Total deaths: {}'.format(total_cases, total_cases))
```

#### Death to Case Ratio

```python
death_rate = total_deaths / total_cases
print('Death to Case ratio: {:0.2f} %'.format(death_rate * 100))
```

#### Total tests, when total test before record was 935310

```python
initial_tests = 935310
total_tests = initial_tests + covid_df.new_tests.sum()
print('Total cases: {}'.format(total_tests))
```

#### Rate of being Test positive

```python
positive_rate = total_cases / total_tests
print('Positive rate: {:0.2f}'.format(positive_rate * 100))
```

#### Days of cases more than 1000

```python
covid_df[covid_df['new_cases'] > 1000]
```

#### Days of higher death rate

```python
covid_df[(covid_df.new_deaths / covid_df.new_cases) > death_rate]
```

#### Days of higher positive rate

```python
covid_df[covid_df.new_cases / covid_df.new_tests > positive_rate]
```

```python
merge_df
```

```python
merge_df['cases_per_million'] = merge_df.new_cases * 1e6 / merge_df.population
merge_df['deaths_per_million'] = merge_df.new_deaths * 1e6 / merge_df.population
merge_df['tests_per_million'] = merge_df.new_tests * 1e6 / merge_df.population
merge_df
```

```python
result_df = merge_df[['date', 'new_cases', 'total_cases', 'new_deaths', 'total_deaths', 'new_tests', 'total_tests', 'cases_per_million', 'deaths_per_million', 'tests_per_million']]
result_df
```

```python
result_df.to_csv('data/results.csv', index=None)
```
