```python
plt.plot(years, oranges)
plt.plot(years, apples)
plt.xlabel('Years')
plt.ylabel('Yield (tons per hactare)');
```

# Chart title and legend

```python
plt.plot(years, oranges)
plt.plot(years, apples)
plt.xlabel('Years')
plt.ylabel('Yield (tons per hectore)')
plt.title('Crop yield')
plt.legend(['orange', 'apple']);
```

# Line marker

```python
plt.plot(years, oranges, marker='o')
plt.plot(years, apples, marker='x')
plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)')
plt.title('crop yield')
plt.legend(['orange', 'apple']);
```

# Styling lines and marker

```python
plt.plot(years, oranges, marker='s', c='r', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, apples, marker='o', c='b', ls='--', lw=3, ms=10, alpha=True)
plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)')
plt.title('crop yield')
plt.legend(['orange', 'apple']);
```

# Scatter plot

```python
import pandas as pd
```

```python
iris_df = pd.read_csv('data/iris.csv')
iris_df
```

```python
iris_df.species.unique()
```

```python
plt.plot(iris_df.sepal_length, iris_df.sepal_width);
```

```python
plt.scatter(iris_df.sepal_length, iris_df.sepal_width);
```

```python
sns.scatterplot(iris_df, x='sepal_length', y='sepal_width', hue='species', s=100);
```

# Customizing seaborn figure

```python
plt.figure(figsize=(12, 6))
plt.title('Sepal Dimentions')
sns.scatterplot(iris_df, x=iris_df.sepal_length, y=iris_df.sepal_width, hue=iris_df.species, s=100);
```

# Histogram

```python
plt.title('Distribution of sepal width')
plt.hist(iris_df.sepal_width);
```

# Controlling size and number of bins

```python
plt.hist(iris_df.sepal_width, bins=5)
plt.title('Distribution of sepal width');
```

```python
import numpy as np

plt.hist(iris_df.sepal_width, bins=np.arange(2, 5, 0.25));
```

# Multiple histograms

```python
setosa_df = iris_df[iris_df.species == 'setosa']
versicolor_df = iris_df[iris_df.species == 'versicolor']
virginica_df = iris_df[iris_df.species == 'virginica']
setosa_df
```

```python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25))
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25))
plt.legend(['setosa', 'versicolor'])
plt.title('Comparing the distribution of sepal width');
```

# Stacked

```python
plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], stacked=True, bins=np.arange(2, 5, 0.25))
plt.title('Stacked')
plt.legend(['setosa', 'versocolor', 'virginica']);
```

# Plotting multiple charts in a grid

```python
fig, axes = plt.subplots(2, 3, figsize=(16, 8))
plt.tight_layout(pad=3)

# Use the axes for plotting
axes[0, 0].plot(years, oranges, 'o-b')
axes[0, 0].plot(years, apples, 's--r')
axes[0, 0].set_xlabel('Year')
axes[0, 0].set_ylabel('Yield (tons per hectare)')
axes[0, 0].set_title('Crop yield')

# Using axes in seaborn
sns.scatterplot(iris_df, x=iris_df.sepal_length, y=iris_df.sepal_width, hue=iris_df.species, ax=axes[0, 1])
axes[0, 1].set_label('Sepal Dimention')

# Using the axes for plotting
axes[0, 2].hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], stacked=True, bins=np.arange(2, 5, 0.25))
axes[0, 2].legend(['setosa', 'versicolor', 'virginica'])

# Using axes in seaborn
sns.barplot(tips_df, x=tips_df.day, y=tips_df.total_bill, hue=tips_df.sex, ax=axes[1, 0])
axes[1, 0].set_title('total bills');

# Using axes in seaborn
sns.heatmap(flight_df, ax=axes[1, 1])
axes[1, 1].set_title('number of passengers(1000s)')

# Using the axes for plotting
axes[1, 2].imshow(img)
axes[1, 2].grid(False)
axes[1, 2].axis('off')
axes[1, 2].set_title('Transformer');
```
